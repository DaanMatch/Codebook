```python
import boto3
import pandas as pd
import io
```

```python
client = boto3.client('s3')
resource = boto3.resource('s3')
my_bucket = resource.Bucket('daanmatchdatafiles')

path = 's3://daanmatchdatafiles/DaanMatch_DataFiles/Final_Data_csr.gov.in.xlsx'
obj = client.get_object(Bucket='daanmatchdatafiles', Key='DaanMatch_DataFiles/Final_Data_csr.gov.in.xlsx')
df = pd.read_excel(io.BytesIO(obj['Body'].read()))
```
